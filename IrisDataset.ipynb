```python

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
from ucimlrepo import fetch_ucirepo





# fetch dataset
iris = fetch_ucirepo(id=53)

# data (as pandas dataframes)
X = iris.data.features
y = iris.data.targets

# metadata
print(iris.metadata)

# variable information
print(iris.variables)

#combine
df = pd.concat([X,y],axis=1)

#rename the columns in order to make easier to read
df.columns = [col.replace('(cm)','') for col in df.columns]
df.rename(columns={'class': 'species'}, inplace=True)

#displaying few rows for a quick check
print("First three rows of the dataset: ")
print(df.head())

#displaying basic stats
print("\nBasic stats of dataset: ")
print(df.describe())

#displaying species
print("\nDistribution of species in the dataset: ")
print(df['species'].value_counts())

#visualizing the relationship betwwen vars by using pairplot
sns.pairplot(df, hue='species')
plt.suptitle("Pairpilot of the Irıs Dataset ",y=1.02)
plt.show()

#plotting histos in order to understand distrubitution pattern
df.hist(edgecolor= 'black', linewidth=1.5, figsize=(15,10))
plt.suptitle("Histograms of features in the Irıs Dataset", y=1.02)
plt.show()

#plotting boxplots for outliers(sepal length/sepal width/petal length/petal width)
plt.figure(figsize=(15,10))
sns.boxplot(x='species', y='sepal length', data=df)
plt.title('Boxplot of Sepal Length by Species ')
plt.show()

plt.figure(figsize=(15,10))
sns.boxplot(x='species', y='sepal width', data=df)
plt.title('Boxplot of Sepal Width by Species ')
plt.show()

plt.figure(figsize=(15,10))
sns.boxplot(x='species', y='petal length', data=df)
plt.title('Boxplot of Petal Length by Species ')
plt.show()

plt.figure(figsize=(15,10))
sns.boxplot(x='species', y='petal width', data=df)
plt.title('Boxplot of Petal Width by Species ')
plt.show()

#calculating to see correlation
print("\nMatrix of Correlation: ")
corr_matrix = df.corr()
print(corr_matrix)

#plotting a heatmap of the matrix of correlation
plt.figure(figsize=(10,5))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title('Heatmap of the Matrix of Correlation ')
plt.show()








```
